```python
import numpy as np
from scipy.constants import *

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import CurrentDistribution as current
```

```python
# Biot-Savart Law
def dB(curr, dL, r_hat, r_mag):
  return (mu_0/(4*pi))*curr*(np.cross(dL,r_hat)/(r_mag**2))
```

```python
radius = 10
dtheta = np.pi/16
loopCurr = current.LoopCurrent(radius,dtheta)
curr_mag = 1
curr = curr_mag*loopCurr.segments

n_steps = 10
vol_x = np.linspace(-radius,radius, num=n_steps, endpoint=True)
vol_y = np.linspace(-radius,radius, num=n_steps, endpoint=True)
vol_z = np.linspace(0,2*radius, num=n_steps, endpoint=True)
vx, vy, vz = np.meshgrid(vol_x, vol_y, vol_z)
vol = np.array([vx.flatten(), vy.flatten(), vz.flatten()]).T
```

```python
B_field = np.empty(vol.shape)

for i in range(0,vol[:].shape[0]):
  r = vol[i]
  r_mag = np.linalg.norm(r)
  r_hat = r/r_mag
  
  B_i = np.empty([curr.shape[0],3])
  
  for j in range(0,curr.shape[0]-1):
    if j < curr.shape[0]:
      dL = curr[j+1] - curr[j]
    else:
      dL = curr[j] - curr[j-1]
    
    B_i[j] = dB(curr_mag, dL, r_hat, r_mag)
  
  B_field[i] = np.sum(B_i)
```
